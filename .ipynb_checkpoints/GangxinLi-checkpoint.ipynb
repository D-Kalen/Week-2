### Profile Analysis

```python
Cab_Data['Date'] = pd.to_datetime(Cab_Data['Date of Travel']-min(Cab_Data['Date of Travel']),unit='D',origin=pd.Timestamp('2016-01-01'))
```

```python
Cab_Data['month'] = pd.DatetimeIndex(Cab_Data['Date']).month
Cab_Data['day'] = pd.DatetimeIndex(Cab_Data['Date']).day
Cab_Data['year'] = pd.DatetimeIndex(Cab_Data['Date']).year
```

```python
Cab_Data['Average Profit Per KM'] = (Cab_Data['Price Charged']- Cab_Data['Cost of Trip'])/Cab_Data['KM Travelled']
```

```python
Cab_Data
```

```python
Labels = ['2016','2017','2018']
Yellow1 = np.mean(Cab_Data[(Cab_Data['year']==2016) & (Cab_Data['Company']=='Yellow Cab')]['Average Profit Per KM'])
Yellow2 = np.mean(Cab_Data[(Cab_Data['year']==2017) & (Cab_Data['Company']=='Yellow Cab')]['Average Profit Per KM'])
Yellow3 = np.mean(Cab_Data[(Cab_Data['year']==2018) & (Cab_Data['Company']=='Yellow Cab')]['Average Profit Per KM'])
Pink1 = np.mean(Cab_Data[(Cab_Data['year']==2016) & (Cab_Data['Company']=='Pink Cab')]['Average Profit Per KM'])
Pink2 = np.mean(Cab_Data[(Cab_Data['year']==2017) & (Cab_Data['Company']=='Pink Cab')]['Average Profit Per KM'])
Pink3 = np.mean(Cab_Data[(Cab_Data['year']==2018) & (Cab_Data['Company']=='Pink Cab')]['Average Profit Per KM'])
Yellow = [Yellow1,Yellow2,Yellow3]
Pink = [Pink1,Pink2,Pink3]

fig, ax = plt.subplots()
width = 0.35

p1 = ax.bar(Labels, Yellow, width, label='Yellow Cab')
p2 = ax.bar(Labels, Pink, width,bottom=Yellow,label='Pink Cab')

ax.set_ylabel('Average Profit Per KM')
ax.set_title('Average Profit Per KM')
ax.set_xlabel('Date of Travel')
ax.bar_label(p1, label_type='center',fmt='%.3f')
ax.bar_label(p2, label_type='center',fmt='%.3f')
ax.legend()
plt.show()
```

```python
Cab_Data['Profit_Yearwise'] = Cab_Data['Cost of Trip']/Cab_Data['Price Charged']
Cab_Data
```

```python
Labels = ['2016','2017','2018','2016 ','2017 ','2018 ']
Yellow1 = np.mean(Cab_Data[(Cab_Data['year']==2016) & (Cab_Data['Company']=='Yellow Cab')]['Profit_Yearwise'])
Yellow2 = np.mean(Cab_Data[(Cab_Data['year']==2017) & (Cab_Data['Company']=='Yellow Cab')]['Profit_Yearwise'])
Yellow3 = np.mean(Cab_Data[(Cab_Data['year']==2018) & (Cab_Data['Company']=='Yellow Cab')]['Profit_Yearwise'])
Pink1 = np.mean(Cab_Data[(Cab_Data['year']==2016) & (Cab_Data['Company']=='Pink Cab')]['Profit_Yearwise'])
Pink2 = np.mean(Cab_Data[(Cab_Data['year']==2017) & (Cab_Data['Company']=='Pink Cab')]['Profit_Yearwise'])
Pink3 = np.mean(Cab_Data[(Cab_Data['year']==2018) & (Cab_Data['Company']=='Pink Cab')]['Profit_Yearwise'])

Yellow = [Yellow1,Yellow2,Yellow3,Pink1,Pink2,Pink3]

fig, ax = plt.subplots()
width = 0.45
bar_colors = ['tab:blue', 'tab:orange', 'tab:red']

# p1 = ax.bar(Labels, Yellow, width, label='Yellow Cab')
# p2 = ax.bar(Labels, Pink, width,bottom=Yellow,label='Pink Cab')
p1 = ax.bar(Labels, Yellow, width, color=bar_colors)
ax.set_ylabel('Profit')
ax.set_title('Profile% Yearwise')
ax.set_xlabel('Pink(Left 3)          Date of Travel          Yellow(Right 3)')
ax.bar_label(p1, label_type='edge',fmt='%.3f')
plt.show()
```
