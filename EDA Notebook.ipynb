```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xlrd
```

```python
df = pd.read_csv("data_cleaning.csv")
df1 = pd.read_csv("city_data_cleaned.csv")
```

```python
df.head()
```

```python
date=[]
for i in df['Date of Travel']:
    xl_date = i
    datetime_date = xlrd.xldate_as_datetime(xl_date, 0)
    date_object = datetime_date.date()
    date.append(date_object)
   
```

```python
df['Date']= date
df
```

```python
#To Delete a whole column
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.drop(['Date of Travel'], axis = 1)
```

```python
#Calling Both Number of Transactions for Both PinkCab and Yellow cab
Yellow_Cab_Transaction = Yellow_Cab.Transaction_ID.count()
Pink_Cab_Transactions = Pink_Cab.Transaction_ID.count()
    
```

```python
#Calling Both profits of the Yellow can and the Pink cab
Yellow_Cab_Profit = Yellow_Cab.Profit_of_Cabs.sum()
Pink_Cab_Profit = Pink_Cab.Profit_of_Cabs.sum()
```

```python
#Total Distance travelled by Both the Yello cab and the Pink Cab
Yellow_Cab_Distance_Trav = Yellow_Cab.KM_Travelled.sum()
Pink_Cab_Distance_Trav = Pink_Cab.KM_Travelled.sum()
```

```python
#Pie Chart for Number of Transactions
labels=['Yellow Cab', 'Pink Cab']
colors = ['#ffff99','#ffb3e6']
plt.pie([Yellow_Cab_Transaction, Pink_Cab_Transactions], labels=labels, colors= colors, autopct='%.2f')

plt.show()

```

FROM THE CHART ABOVE WE CAN CLARLY SEE THAT THE NUMBER OF TRANSACTIONS FOR THE YELLOW CAB SURPASES THAT FOR THE PINK CAB, MEANING THAT CUSTOMERS ENGAGE MORE WITH THE SERVICES OF THE YELLOW CAB THAN THE PIN CAB.

```python
#Pie Chart for Total Profits For Yellow And Pink Cab
labels=['Yellow Cab', 'Pink Cab']
colors = ['#ffff99','#ffb3e6']
plt.pie([Yellow_Cab_Profit, Pink_Cab_Profit], labels=labels, colors= colors, autopct='%.2f')

plt.show()

```

THIS CHART GIVES US THE TOTAL PROFIT FROM THE YEAR 2016 - 2018 AND WE CAN CLEARLY SEE A WOOPING 88.89 PERCENT OF THE TOTAL PROFIT THAT BELONGS TO THE YELLOW CAB, AND CLEARLY THE HUGE AMOUNTS OF TRANSACTIONS CAN ACCOUNT FOR THIS. ONCE AGAIN THE PINK CAB'S PROFIT IS NOT REALLY ENCOURAGING.

```python
#Pie Chart for Total Distance travelled by Both the Yello cab and the Pink Cab
labels=['Yellow Cab', 'Pink Cab']
colors = ['#ffff99','#ffb3e6']
plt.pie([Yellow_Cab_Distance_Trav, Pink_Cab_Distance_Trav], labels=labels, colors= colors, autopct='%.2f')

plt.show()
```

FROM THE PIE CHART ABOVE, IT IS CLEAR THAT 74.44 PERCENT OF TOTAL DISTANCE COVERED, BELONGS TO THE YELLOW CAB. THIS IMPLIES THAT CUSTOMERS ACTIVELY SEEK THE SERVICES OF THE YELOOW CAB COMPARED TO THE PINK CAB

## RECOMENDATION


We have evaluated both the cab companies on following points and found Yellow cab better than Pink cab:
• Customer Reach : Yellow cab has higher customer reach compared to the Pink cab. We have
also observed that Yellow cab is doing good in covering other cab users as compared to Pink cab.
• Customer Retention: From the data we Analysed, we found that Yellow Cab has a better customer retention as compared to the Pink Cab.We also noticed some customers switching from the Pink cab to the Yellow cab.
• Average Profit per KM: Yellow cab’s average profit per KM is almost three times the average profit per KM of the Pink cab.
• Income wise Reach :Both the cabs are very popular in high and medium income class but here also Yellow cab is performing better
than Pink cab in offering their services to all the three income class group (low, medium and high)


On the basis of above point , we will recommend Yellow cab for investment.
